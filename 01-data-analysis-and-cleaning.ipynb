#### Import Libraries

```python
import numpy as np
import pandas as pd
```

#### Import Dataset

```python
df = pd.read_csv("diabetic.csv")
```

#### Analyze

```python
df.info()
```

```python
for column in df.columns:
    print(column)
    print(*sorted(df[column].unique()))
```

```python
# Handle Glucose column
df['Glucose'] = df['Glucose'].replace(0, np.nan)
median = df['Glucose'].median()
df['Glucose'] = df['Glucose'].fillna(median)
```

```python
# Handle BloodPressure Column
df['BloodPressure'] = df['BloodPressure'].replace(0, np.nan)
median = df['BloodPressure'].median()
df['BloodPressure'] = df['BloodPressure'].fillna(median)
```

```python
# Handle SkinThickness Column
df['SkinThickness'] = df['SkinThickness'].replace(0, np.nan)
median = df['SkinThickness'].median()
df['SkinThickness'] = df['SkinThickness'].fillna(median)
```

```python
# Handle Insulin Column
df['Insulin'] = df['Insulin'].replace(0, np.nan)
median = df['Insulin'].median()
df['Insulin'] = df['Insulin'].fillna(median)
```

```python
# Handle BMI Column
df['BMI'] = df['BMI'].replace(0, np.nan)
median = df['BMI'].median()
df['BMI'] = df['BMI'].fillna(median)
```

```python
# save as csv
df.to_csv('cleaned_data.csv', index = False)
```
